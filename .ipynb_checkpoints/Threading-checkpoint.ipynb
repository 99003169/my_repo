```python
#separate threads for separate functions


import threading
import time

def task_one():
    print("A")
    for x in range(1,100):
        print("A: ", x)
        time.sleep(1)

def task_two():
    print("B")
    for x in range(1,100):
        print("B: ", x)
        time.sleep(1)
    
if __name__ == '__main__':
    t1 = threading.Thread(target = task_one, name = "A")             #separate threads for separate functions
    t2 = threading.Thread(target = task_two, name = "B")
    
    t1.start()          # start thread
    t2.start()
    
    t1.join()          # wait until the thread terminates to execute following statements
    t2.join()
    
    print("thank you")
```

```python
# data driven parallelism


import threading
import time

def task_common(mystr):      #common function
    print("Thread")
    for x in range(1,100):
        print(mystr,":", x)
        time.sleep(1)

    
if __name__ == '__main__':
    t1 = threading.Thread(target = task_common, args = ("A",))   # data driven parallelism
    t2 = threading.Thread(target = task_common, args = ("B",))     # different data
    t3 = threading.Thread(target = task_common, args = ("C",))
    
    t1.start()
    t2.start()
    t3.start()
    
    t1.join()
    t2.join()
    t3.join()
    
    print("thank you")
```

```python
#target is a callable object


import threading
import time

class mytask:
    def __call__(self, mystr):
        print("Thread")
        for x in range(1,100):
            print(mystr,"--", x)
            time.sleep(1)

    
if __name__ == '__main__':
    t1 = threading.Thread(target = mytask(), args = ("A",))     #target is an object
    t2 = threading.Thread(target = mytask(), args = ("B",))
    t3 = threading.Thread(target = mytask(), args = ("C",))
    
    t1.start()
    t2.start()
    t3.start()
    
    t1.join()
    t2.join()
    t3.join()
    
    print("thank you")
```

```python
#user defined class inheriting threading.Thread()


import threading
import time

class mythread(threading.Thread):
    def __init__(self, mystr):
        super().__init__()
        self.label = mystr
    def run(self):
        print("Thread", self.label)
        for x in range(1,100):
            print(self.label,"--", x)
            time.sleep(1)

    
if __name__ == '__main__':
    t1 = mythread("A")   
    t2 = mythread("B")
    t3 = mythread("C")
    
    t1.start()
    t2.start()
    t3.start()
    
    t1.join()
    t2.join()
    t3.join()
    
    print("thank you")
```

```python
'''race condition - concurrent updates to shared resources will lead to 
inconsistancy / resources will be corrupted '''

'''Mutual exclusion :: two or more function shoudnt execute critical section together'''

'''locks / semaphores for mutual exclusion'''

import threading
import time

val =  100
def task_one():
    global val
    print("A")
    for x in range(1,1000):
        val = val + 1               #critical section
        print("A", val)

def task_two():
    global val
    print("B")
    for x in range(1,1000):
        val = val - 1               #critical section
        print("B", val)
    
if __name__ == '__main__':
    t1 = threading.Thread(target = task_one, name = "A")
    t2 = threading.Thread(target = task_two, name = "B")
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("thank you")
```

```python
#global object for Lock()


import threading
import time

val =  100

mylock = threading.Lock()               #global object

def task_one():
    global val
    print("A")
    mylock.acquire()
    for x in range(1,100):
        val = val + 1               #critical section
        print("A", val)
        time.sleep(1)
    mylock.release()

def task_two():
    global val
    print("B")
    mylock.acquire()
    for x in range(1,100):
        val = val - 1               #critical section
        print("B", val)
        time.sleep(1)
    mylock.release()
    
if __name__ == '__main__':
    t1 = threading.Thread(target = task_one, name = "A")
    t2 = threading.Thread(target = task_two, name = "B")
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("thank you")
```

```python
#local object for Lock()


import threading
import time

val =  100

def task_one(mylock):
    global val
    print("A")
    mylock.acquire()
    for x in range(1,100):
        val = val + 1               #critical section
        print("A", val)
        time.sleep(1)
    mylock.release()

def task_two(mylock):
    global val
    print("B")
    mylock.acquire()
    for x in range(1,100):
        val = val - 1               #critical section
        print("B", val)
        time.sleep(1)
    mylock.release()
    
if __name__ == '__main__':
    mylock = threading.Lock()          #local object
    t1 = threading.Thread(target = task_one, args = (mylock,))
    t2 = threading.Thread(target = task_two, args = (mylock,))
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("thank you")
```

```python
# Example for threading.Lock()


class mydata:
    def __init__(self, val):
        self.val = val
    def inc(self):
        self.val = self.val + 1
    def dec(self):
        self.val = self.val - 1
    def printval(self):
        print(self.val)

def task_one(mylock,myobj):
    with mylock:                               #acquire & release
        for x in range(1,1000):
            myobj.inc()
            myobj.printval()
        
def task_two(mylock,myobj):                          #acquire & release
    with mylock:
        for x in range(1,1000):
            myobj.dec()
            myobj.printval()
                
if __name__ == '__main__':
    mylock = threading.Lock()
    myobj = mydata(100)
    t1 = threading.Thread(target = task_one, args = (mylock,myobj))
    t2 = threading.Thread(target = task_two, args = (mylock,myobj))
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("thank you")
```

```python
# semaphore example


import threading
import time

val = 100

sem = threading.Semaphore(value=0)    # value=0 block and add to queue , value>0 go ahead

def task_one(mylock):  
    global val
    sem.acquire()
    with mylock:
        for x in range(1,100):
            val = val + 1
            print("A: ", val)
            time.sleep(1)

def task_two(mylock):
    global val
    with mylock:
        for x in range(1,100):
            val = val - 1
            print("B: ", val)
            time.sleep(1)
    sem.release()
    
if __name__ == '__main__':
    mylock = threading.Lock()
    t1 = threading.Thread(target = task_one, name = "A", args = (mylock,))
    t2 = threading.Thread(target = task_two, name = "B", args = (mylock,))
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
    print("thank you")
```

```python
import threading


class StackFullError(Exception):
    def __init__(self, msg):
        self.msg = msg

    def __str__(self):
        return self.msg


class StackEmptyError(Exception):
    def __init__(self, msg):
        self.msg = msg

    def __str__(self):
        return self.msg


class Stack:
    def __init__(self, len):
        self.mylist = []
        self.maxlen = len

    def push(self, val):
        if len(self.mylist) == self.maxlen:
            raise StackFullError("Overflow!")        
        self.mylist.append(val)

    def pop(self):
        if len(self.mylist) == 0:
            raise StackEmptyError("Underflow!")
        return self.mylist.pop()


if __name__ == "__main__":
    s1 = Stack(5)
    try:
        s1.push(11)
        s1.push(12)
        s1.push(13)
        s1.push(14)
        s1.push(15)
        s1.push(16)
        print(s1.pop())
    except StackEmptyError as e:
        print("Stack is Empty.", e)
    except StackFullError as e:
        print("Stack is Full.", e)
```

```python
import threading
import random

sem = threading.Semaphore(value = 1)

class MyArray:
    def __init__(self, len):
        self.data = []
        self.maxlen = len
    
    def populate(self, loc):
        sem.acquire()
        with loc:
            print("populate - start")
            for i in range(0, self.maxlen):
                self.data.append(random.randint(1, 100))
            print("populate - done")

    def printsum(self, loc):                     #sequential sum
        print("printsum")
        total = 0
        for x in range(0, self.maxlen):
            total += self.data[x]
        with loc:
            print("prinstsum: ", total)

    def sub_total(self, loc, start, end):        # parallel_sum
        print("sub_total")
        total = 0
        for x in range(start, end):
            total += self.data[x]
        with loc:
            print("sub_total: ", total)
        sem.release()

if __name__ == "__main__":
    loc = threading.Lock()
    arr = MyArray(100)
    
    t1 = threading.Thread(target = arr.populate, name = "A", args = (loc,))
    t2 = threading.Thread(target = arr.printsum, name = "B", args = (loc,))
    t3 = threading.Thread(target = arr.sub_total, name = "C", args = (loc, 0, 100))
    t4 = threading.Thread(target = arr.sub_total, name = "D", args = (loc, 0, 10))
    t5 = threading.Thread(target = arr.sub_total, name = "E", args = (loc, 0, 20))
    
    t1.start()
    t3.start()
    t2.start()
    t4.start()
    t5.start()

    t1.join()
    t2.join()
    t3.join()
    t4.join()
    t5.join()
    
    print("Done")
```

```python
import random
import threading
import time


class MyLogger:
    def __init__(self, path):
        self.fd = open(path, "w")     # may add with clause

    def log(self, msg):
        self.fd.write(msg)

    def __exit__(self):
        self.fd.close()


maxTimes = 10


def task_one(logger,loc):
    for k in range(0, maxTimes):
        tval = random.randint(18, 40)
        with loc:
            logger.log("Temperature is :" + str(tval) + "\n")
            time.sleep(1)


def task_two(logger,loc):
    for k in range(0, maxTimes):
        hval = random.randint(50, 100)
        with loc:
            logger.log("Humidity is :" + str(hval) + "\n")
            time.sleep(1)


if __name__ == "__main__":
    loc = threading.Lock()
    logger = MyLogger("sample.txt")
    t1 = threading.Thread(target=task_one, args=(logger, loc,))
    t2 = threading.Thread(target=task_two, args=(logger, loc,))

    t1.start()
    t2.start()

    t1.join()
    t2.join()
    logger.__exit__()
    
print("Done")
```

```python
"""
Using condition variables for synchronization between threads
"""
import threading
import time


def task_one(cv, mlock):
    time.sleep(5)
    mlock.acquire()
    # cv.wait()                #release mlock temporarily and acquire mlock post notify
    print("Task A")
    for x in range(1, 10):
        print("A--", x)
        time.sleep(1)
    cv.notify()
    mlock.release()


def task_two(cv, mlock):
    mlock.acquire()
    print("Task B")
    cv.wait()
    for x in range(1, 10):
        print("B--", x)
        time.sleep(1)
    #cv.notify()
    mlock.release()


if __name__ == "__main__":

    mlock = threading.Lock()
    cv = threading.Condition(mlock)

    t1 = threading.Thread(target=task_one, args=(cv, mlock,))
    t2 = threading.Thread(target=task_two, args=(cv, mlock,))

    t1.start()
    t2.start()

    t1.join()
    t2.join()
    
    print("Done")
```
